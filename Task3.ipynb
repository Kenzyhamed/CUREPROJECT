```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
RecentOrders = pd.read_excel('RecentOrders.xlsx')
assorted= RecentOrders.sort_values('Status', ignore_index=True)
assorted.Status.value_counts()
f1, f0 =assorted.Status.value_counts()
fig=plt.figure(figsize=(15,2.5))
axs1=plt.subplot2grid(shape=(1,4), loc=(0,0))
axs2=plt.subplot2grid(shape=(1,4), loc=(0,1))
axs3=plt.subplot2grid(shape=(1,4), loc=(0,2))
axs4=plt.subplot2grid(shape=(1,4), loc=(0,3))
plt.tight_layout()

plt.tight_layout()

axs1.hist(assorted.iloc[0:f0,1], edgecolor='b',fc='none', label='c0(In progress)')
axs1.hist(assorted.iloc[f0:f0+f1,1], edgecolor='r',fc='none', label='c1(Shipped)')
axs1.set_xlabel('Quantity of Items Ordered')
axs1.set_ylabel('Frequency')
axs1.legend()

axs2.scatter(assorted.iloc[0:f0,0],assorted.iloc[0:f0,2], label='c0(IP)')
axs2.scatter(assorted.iloc[f0:f0+f1,0],assorted.iloc[f0:f0+f1,2], label='c1(SH)')
axs2.set_xlabel('Price per Unit(dollars)')
axs2.set_ylabel('Sales(dollars)')
axs2.legend()

axs3.scatter(assorted.iloc[0:f0,0],assorted.iloc[0:f0,1], label='c0(IP)')
axs3.scatter(assorted.iloc[f0:f0+f1,0],assorted.iloc[f0:f0+f1,1], label='c1(SH)')
axs3.set_xlabel('Price per Unit(dollars)')
axs3.set_ylabel('Qty. of items ordered')
axs3.legend()

axs4.scatter(assorted.iloc[0:f0,1],assorted.iloc[0:f0,2], label='c0(IP)')
axs4.scatter(assorted.iloc[f0:f0+f1,1],assorted.iloc[f0:f0+f1,2], label='c1(SH)')
axs4.set_xlabel('Qty. of items ordered')
axs4.set_ylabel('Sales(dollars)')
axs4.legend()
plt.show()
```

graph 2 is a scatter graph that compares the price of one unit in the order to the total amount of sales made from that order.
graph 3 is a scater graph that compares the price of one unit in the order to the amount ordered (amount in the bulk thats ordered)
graph 4 is a scatter graph that compares the Quantity of items ordered to the total amount of sales made from that order.
